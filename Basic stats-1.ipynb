###  Standardization of Numerical Variables

#### ●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

Explaination:

Explaining the concept of Standardization (Z-Score Normalization):

Standardization (Z-score normalization) transforms the data into standard value which hepls for the machine learning model to learn from the data more easily. The standardized valued lies between 0 and 1, mean (μ) of 0 and a standard deviation (σ) of 1. All the values lies between this range. We must remember that the data of the columns that are standardized are also equivalent the data that is not standardized.

z = (x - μ)/ σ

where, x = the original data point. μ = the mean of the data. σ = the standard deviation of the data.

Standardizing the numerical columns using the formula: 
z=x-mu/σ

```python
df.info()
```

```python
df_1=df
df_1.info()
```

```python
# Standardization of Volume Column 

mean = df_1['Volume'].mean()   #Find mean
std = df_1['Volume'].std()   #Find standard deviation

# applying formula z = (x - mu) / sigma
df_standardized = (df_1['Volume']-mean)/std

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.histplot(data=df_1['Volume'], kde=True, ax=axes[0])    # Original data distribution
axes[0].set_title("Original Data Distribution of Volume")

sns.histplot(data=df_standardized, kde=True, ax=axes[1])   # Standardized data distribution
axes[1].set_title("Standardized Data Distribution of Volume")

df_1['Volume'] = df_standardized         

plt.tight_layout()
plt.show()
```

######  Here we can observe that only scale of the data is reduced but the distribution across the plane is the same.

Hence we can say the data is standardized

## Conversion of Categorical Data into Dummy Variables

#### Step 1:

Objective: To transform categorical variables into a format that can be provided to ML algorithms.

Discussing the need for converting categorical data into dummy variables :

The most important reason to convert categorical data into numerical is that our computers don't understand categorical data, the only type of data it understands is numerical. Although it can understand after converting categorical to numerical data, we must keep in mind that our main goal is to train the model so it is best practice to avoid any unnecessary clutter in the data and stick to numerical data.

Technique used for doing this is called OneHotEncding.

Reasons to do OneHotEncoding-

1.Compatibility with Algorithms: Most of the algorithms take data in numerical format so it is easily compatible.
2.Allows mathematical operations: Since the data is in numerical format, we can perform add, sub, mul and etc...
3.Representation of Relationships: Proper encoding and decoding can be done
4.Model Performance: Increases the preformace of model since there is no need to do unnecessary convertions for understanding

#### Step 2:

Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
